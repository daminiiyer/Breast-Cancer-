```python
# importing libraries to use
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
df = pd.read_csv("breast cancer.csv", encoding='latin1') 
df.head(5)
```

```python
# dropping null columns
nulls =df.isnull().sum()
print("Total number of null values: ", nulls.sum())
df = df.drop(columns=['Unnamed: 32'])
df.head(5)
```

```python
#checking if the null columns have been dropped
nulls =df.isnull().sum()
print("Total number of null values: ", nulls.sum())
```

```python
#encoding categorical data into numeric data to be interpreted 
#by the machine learnig model.
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})
```

```python
# selecting columns for x and y.
# X- columns used to predict output. (feature data)
# Y- column showing output. (target data)
selected_columns = ['radius_mean','texture_mean','concave points_mean','radius_se','compactness_worst','concavity_worst','concave points_worst']
X = df[selected_columns]   
y = df['diagnosis']
```

```python
# splitting x and y into training data and testing date depending on 
# specifications in the bracket. Here, its 40% test and 60% train.
# model sees both x_train and y_train, but only sees x_test and not y_test.
# it predicts first, and then sees y_test to check its accuracy.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4,train_size=0.6, random_state=42)
```

```python
print(f'Shape of X: {X.shape}')
print(f'Shape of y: {y.shape}')
```

```python
#scaling
scaler = StandardScaler() #created an instance of StandardScaler
# transforms data such that mean=0 (all vals centered around 0
# and std deviation=1. (vals are within 1 unit of mean)
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#logistic regression: supervised ml algo used to predict one of two 
# possibilities (binary classification).
# the sigmoid function converts the linear graph (randing from 0 to 1)
# into an S curve by converting it into a probability using the 
 # sigmoid function formula.if probability>0.5, classifies as 1, else 0.
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
model = LogisticRegression(random_state=42)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
#accuracy: no. of correct predictions/total no. of predictions
# confusion matrix: TN  FP
#                   FN  TP
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

cm = confusion_matrix(y_test, y_pred)
print(f'Confusion Matrix:\n{cm}')

report = classification_report(y_test, y_pred)
print(f'Classification Report:\n{report}')
```
